### P6. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

````python
g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]

index = np.arange(5)
y = ['M', 'T', 'W', 'Th', 'F']
bar_width = 0.20

plt.bar(index, g1, bar_width, color = 'C0', label = 'Group1')
plt.bar(index + bar_width, g2, bar_width, color = 'C1', label = 'Group2')
plt.bar(index + bar_width*2 , g3, bar_width, color = 'C2', label = 'Group3')
plt.xticks(index + bar_width, y)
plt.xlabel('Groups')
plt.ylabel('Productivity')
plt.title('Productivity during week')
plt.legend()

plt.show()
````
